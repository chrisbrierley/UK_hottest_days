```python
import xarray as xr
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
UKGrid=xr.open_dataset("UKGrid/maxtmax-uk.nc")
```

```python
this_year_date=pd.DatetimeIndex(data=['2022-01-01T00:00:00.000000000'])
this_year=xr.Dataset(
    {
        "tasmax": (38.1),
    },
    coords={"time": this_year_date},
)
```

```python
warmest_day_of_year=xr.combine_by_coords([UKGrid,this_year])
warmest_day_of_year
```

```python
plt=warmest_day_of_year.tasmax.plot(ylim=[25,45],xlim=pd.DatetimeIndex(['1960-01-01','2080-01-01']),marker='o', color='black')
```

```python
UKCP18=xr.open_mfdataset(paths="UKCP18/*.nc")
UKCP18.maxtmax[:,0].plot(ylim=[25,50])
UKCP18.time
```
